# sort

```python
# sort

# sort_index()          index e baglı olarak sıralar
# sort_values()         istenilen sütundaki değerlere göre sıralar

# istenilen sütun yada index e baglı olarak DataFrame i sıralar. yani columnx için sort yapılacaksa, 
# columnx kucukten buyuge sıralanır diğer sütunlarda columnx e göre ayarlanır.
# kalıcı işlem yapmaz. kalıcı olması için inplace=True kullanılmalı
```

```python
df5=pd.DataFrame(np.random.randn(10,2),index=[1,4,6,2,3,5,9,8,0,7],columns=['col2','col1'])
```

```python
df5
```

```python
df5.sort_index()                                 # index e gore kucukten buyuge sıralar           
```

```python
df5.sort_index(ascending=False)                  # ascending=False ile buyukten kucuge sıralar
```

```python
df5.sort_index(axis=1)                          # axis=1 ile column isimlerine göre sıralar
```

```python
df5.sort_values(by="col2")                            # col2 ye göre sıraladı 
```

```python
df5.sort_values(by="col1", ascending=False)           # col1 buyukten kucuge sıralandı
```

```python
df5.sort_values(by="col1", ascending=False, ignore_index=True)       # col1 e göre sıralar index leri 0 dan başlayarak sıralar
                                                                     # yeni index atanmış olur
```

```python
df6 = pd.DataFrame({'col1':[2,1,1,1],'col2':[1,3,2,4],'col3':[5,6,7,8]})
```

```python
df6
```

```python
df6.sort_values(by=["col1","col2"])          # once col1 e bakarak sıralar. col1 de aynı değerler varsa 
                                             # bunları col2 değerlerine göre sıralar
```
