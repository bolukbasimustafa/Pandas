```python
# merge

# ortak bir sütun üzerinden merge yapılabilir
```

```python
x = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
y = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
x
```

```python
y
```

```python
pd.merge(x,y, how="inner", on="key")            # DataFrame lerin key sütununda benzer olan değerleri alır. inner kesişimdir
```

```python
pd.merge(x,y,how="outer",on="key")               # DataFrame lerin key sütunundaki tüm değerleri alır. outer kesişimdir
```

```python
pd.merge(x,y,how="left",on="key")       # DataFrame lerden solda yazılan x DataFrame inin key sütununda yer alan değerleri alır
```

```python
pd.merge(x,y,how="right",on="key")      # DataFrame lerden sağda yazılan y DataFrame inin key sütununda yer alan değerleri alır
```

```python
# merge yapılacak ortak bir sütun yoksa aşağıdaki yöntem kullanılır
```

```python
xx = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x'],
                    'lvalue': [2, 3, 5, 7]})
yy = pd.DataFrame({'rkey': ['a', 'x', 'z', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
xx
```

```python
yy
```

```python
pd.merge(xx,yy,left_on="lkey",right_on="rkey", how="outer")
```

```python
pd.merge(xx,yy,left_on="lkey",right_on="rkey", how="left")
```

```python
pd.merge(xx,yy,left_on="lkey",right_on="rkey", how="right")
```

```python
pd.merge(xx,yy,left_on="lkey",right_on="rkey", how="inner")
```

```python
aa = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index = ['K0', 'K1', 'K2'])
bb = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index = ['K0', 'K2', 'K3'])
```

```python
aa
```

```python
bb
```

```python
aa.join(bb)                          # aa DataFrame i yazılır ve bb DataFrame inde yer alan aynı indexlerin karşılığı yazılır
```

```python
bb.join(aa)                         # bb DataFrame i yazılır ve aa DataFrame inde yer alan aynı indexlerin karşılığı yazılır
```

```python
aa.join(bb, how="outer")            # her iki DataFrame de yer alan tüm indexler yazılır ve karşılıkları varsa yazılır yoksa nan
```

```python
aa.join(bb, how="inner")            # her iki DataFrame de yer alan ortak indexler ve value leri yazılır 
```
